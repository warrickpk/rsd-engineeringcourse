```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def firstderivative(eps):
    def differentiate(func):
        def _helper(x):
            return (func(x+eps) - func(x)) / eps
        return _helper
    return differentiate
```

```python
@firstderivative(eps=0.0001)
def my_func(x):
    return math.sin(x)
```

```python
%matplotlib inline
plt.plot(list(map(my_func, np.arange(-10,10,0.01))))
plt.plot(list(map(lambda x: math.sin(x), np.arange(-10,10,0.01))))
```

```python
def func(x):
    return x**4

func_deriv = firstderivative(0.00001)
func_deriv = func_deriv(func)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
func_helper = np.vectorize(func)
func_deriv_helper = np.vectorize(func_deriv)
plt.plot(func_helper(np.arange(-10,10,0.01)))
plt.plot(list(map(func_deriv,np.arange(-10,10,0.01))))
```
